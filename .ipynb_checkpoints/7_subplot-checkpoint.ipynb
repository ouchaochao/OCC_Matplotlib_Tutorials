```python
# 一张图有多个小图
import matplotlib.pyplot as plt
```

```python
plt.figure()

# 行；列；第几个图
plt.subplot(221)
plt.plot([0,1],[0,1])

plt.subplot(222)
plt.plot([0,1],[0,2])

plt.subplot(223)
plt.plot([0,1],[0,3])

plt.subplot(224)
plt.plot([0,1],[0,4])

plt.show()
```

```python
plt.figure()

# 行；列；第几个图
plt.subplot(211)
plt.plot([0,1],[0,1])

# 注意4，这是两行三列的，第二行开始就是第4个图了
plt.subplot(234)
plt.plot([0,1],[0,2])

plt.subplot(235)
plt.plot([0,1],[0,3])

plt.subplot(236)
plt.plot([0,1],[0,4])

plt.show()
```

```python
# 6个小图 法一：subplot2grid
plt.figure()
# 三行三列，从0,0位置开始
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3, rowspan=1)
ax1.plot([1,2], [1,2])
ax1.set_title('ax1_title')

ax2 = plt.subplot2grid((3,3), (1,0), colspan=2,)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2,)
ax4 = plt.subplot2grid((3,3), (2,0),)
ax5 = plt.subplot2grid((3,3), (2,1),)

plt.tight_layout()
plt.show()
```

```python
# 法二：gridspec
import matplotlib.gridspec as gridspec
plt.figure()
gs = gridspec.GridSpec(3,3)
ax1 = plt.subplot(gs[0,:])
ax2 = plt.subplot(gs[1,:2])
ax3 = plt.subplot(gs[1:3,2])
ax4 = plt.subplot(gs[2,0])
ax5 = plt.subplot(gs[2,1])

plt.tight_layout()
plt.show()
```

```python
# 法三
f, ((ax11,ax12),(ax21,ax22)) = plt.subplots(2, 2, sharex=True, sharey=True)
ax11.scatter([1,2],[1,2])
plt.tight_layout()
plt.show()
```

```python
# 图中图
fig = plt.figure()
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 8, 6, 5]

left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('ax1')

left, bottom, width, height = 0.2, 0.6, 0.2, 0.2
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('ax2')

left, bottom, width, height = 0.6, 0.15, 0.2, 0.2
ax3 = fig.add_axes([left, bottom, width, height])
ax3.plot(x, y, 'r')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('ax3')
```

```python
# 次坐标轴
import numpy as np
x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -y1

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b--')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1', color='g')
ax2.set_ylabel('Y2', color='b')
```
